```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  

df = pd.read_json("08_pandas_aula_02.json")
print(df)
print(df.dtypes)
```

```python
df["mes"] = pd.to_datetime(df["mes"])
print(df.dtypes)
```

```python
df = df.set_index("mes").sort_index()
print(df.head())
```

```python
ax = df.plot(figsize=(8,4),title="Evolucao mensal de Visualizacoes por serie")
ax.set_xlabel("mes")
ax.set_ylabel("numero de visualizacoes")
plt.show()
```

```python
totais = df.sum(axis=0).sort_values(ascending=False)
ax = totais.plot(kind="bar", rot=45, title="totais de visualizacoes por series ao longo do ano")
ax.set_xlabel("Serie")
ax.set_ylabel("Total de visualizacoes")
plt.show()
```

```python
medias = df.mean(axis=0).sort_values(ascending=False)
ax = medias.plot(kind="barh", title="Media de visualizacoes por serie")
ax.set_xlabel("Media")
ax.set_ylabel("Serie")
plt.show()
```

```python
top3 = totais.head(3)
ax = top3.plot(kind="pie", autopct="%1.1f%%", ylabel="", title="Participacao das top 3 series")
plt.show()
```

```python
ax = df["Dexter"].plot(kind="hist", bins=5, title="Distribuicao de visualizacoes - Dexter")
ax.set_xlabel("Visualizacoes")
plt.show()
```

```python
ax=df.plot(kind="box", title="Boxplot das visualizacoes por Serie")
plt.show()
```

```python
ax = df.plot(kind='scatter', x = "Breaking Bad", y = "Dexter", title="Sem sentido, como a Vida")
plt.show()
```
